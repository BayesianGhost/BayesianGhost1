```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
```

```python
data0=pd.read_csv('/home/sam/Documents/pima-indians-diabetes.csv')
data0.info()
```

```python
data=data0.to_numpy()
x=torch.tensor(data[:,0:8],dtype=torch.float32)
y=torch.tensor(data[:,8],dtype=torch.float32).reshape(-1,1)

xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size =.12, random_state = 42)
```

```python
model = nn.Sequential(
    nn.Linear(8, 12),
    nn.ReLU(),
    nn.Linear(12, 8),
    nn.ReLU(),
    nn.Linear(8, 1),
    nn.Sigmoid())
```

```python
n_epochs = 100
batch_size = 15

loss_fn = nn.BCELoss()  # binary cross entropy
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

```python
for epoch in range(n_epochs):
    for i in range(0, len(xtrain), batch_size):
        Xbatch = xtrain[i:i+batch_size]
        y_pred = model(Xbatch)
        ybatch = ytrain[i:i+batch_size]
        loss = loss_fn(y_pred, ybatch)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f'Finished epoch {epoch}, latest loss {loss}')
```

```python
with torch.no_grad():
    y_pred = model(xtest)

accuracy = (y_pred.round() == ytest).float().mean()
print(f"Accuracy {accuracy}")
```
